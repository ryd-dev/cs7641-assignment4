```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def rew_delta(df, mdp, learner):
    f, (ax) = plt.subplots(1, 1)
    ax.set_title(f'{mdp}: {learner} - Value and Delta vs Episodes')
    ax.set_xlabel('Episodes')
    ax.set_ylabel('Value')

    lns1 = ax.plot(df.index.values, df['reward'], linewidth=1, label='Value')

    ex_ax = ax.twinx()
    lns2 = ex_ax.plot(df.index.values, df['delta'], linewidth=1, label='Delta', color='orange')
    ex_ax.set_ylabel('Delta')
    ex_ax.tick_params('y')

    ax.grid()
    ax.axis('tight')

    f.tight_layout()

    lns = lns1 + lns2
    labs = [l.get_label() for l in lns]
    ax.legend(lns, labs, loc=0)
```

```python
cwvi = pd.read_csv('output/report/VI/cliff_walking_0.3.csv')
rew_delta(cwvi, 'Cliff Walking', 'Value Iteration')
```

```python
cwpi = pd.read_csv('output/report/PI/cliff_walking_0.7.csv')
rew_delta(cwpi, 'Cliff Walking', 'Policy Iteration')
```
